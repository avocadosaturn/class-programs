```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the two-column data (q, f(q))
q, f = np.loadtxt("psd_FD_single.dat", unpack=True)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(q, f, marker='o', markersize=3, linewidth=1)
plt.xlabel(r'$q = p/T$')
plt.ylabel(r'probability of occupancy')
plt.title('Fermi-Dirac Phase-Space Distribution')
plt.tight_layout()
plt.savefig("psd_FD_single.png", dpi=150)
plt.show()
```
